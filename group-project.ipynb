```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get raw data
url = "https://raw.githubusercontent.com/Elmylmi/gp-project/main/data/mi_health_cannabis_netherlands_usd_edited.csv"
data_cannabisUsers  = pd.read_csv(url)
data_cannabisUsers.head()
```

```python
data_cannabisUsers = data_cannabisUsers.iloc[25:29]
print(data_cannabisUsers)
data_cannabisUsers.set_index(data_cannabisUsers['Name'], inplace=True)
data_cannabisUsers = data_cannabisUsers.drop(['Region', 'Market', 'Chart', 'Name', 'Unit', 'Source'], axis=1)
data_cannabisUsers = data_cannabisUsers.T

data_cannabisUsers
```

```python
# Turn into plot
data_cannabisUsers.plot()
plt.title('GDP in Canada and Netherlands')
plt.ylabel('GDP in Billion, $US')
plt.xlabel('Year')
plt.legend()
plt.show()
```
